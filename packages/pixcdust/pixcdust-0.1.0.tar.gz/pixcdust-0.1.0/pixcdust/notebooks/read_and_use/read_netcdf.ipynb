# Simply read a single PixC netcdf


```python
from pixcdust.readers.netcdf import PixCNcSimpleReader
```

```python
path = "/tmp/pixc/SWOT_L2_HR_PIXC_011_148_098R_20240220T032331_20240220T032342_PIC0_01.nc"
ncsimple = PixCNcSimpleReader(path)#, ['height', 'sig0', 'classification'])

```

```python
ncsimple.open_dataset()
ncsimple.data
```

```python
# show some information
_, dt_time_start, cycle_number, pass_number, tile_number = (
    ncsimple.extract_info_from_nc_attrs(path)
)
print(dt_time_start, cycle_number, pass_number, tile_number)
```

## use data with xarray

```python
ncsimple.to_xarray() # same thing as ncsimple.data
```

## use data with pandas.dataframe

```python
ncsimple.to_dataframe()
```

## use data with geopandas.geodataframe

```python
gdf = ncsimple.to_geodataframe()
gdf
```
